# Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#create a boxplot to visualize distribution/ spread of values
```

```python
plt.figure(figsize=(6,1))
plt.title("Distribution of values")
sns.boxplot(data['Price'])
```

```python
plt.figure(figsize=(5,3))
plt.title('Histograms of The Prices')
sns.histplot(data['Price'], bins=range(0,(10**7+1), 10**6))
```

```python
plt.figure(figsize=(5,1))
plt.title('Distribution of mileage')
sns.boxplot(x=data['Mileage'])
```

```python
plt.figure(figsize=(5,1))
plt.title('Distribution of mileage')
sns.histplot(data['Mileage'], bins=range(0,500000,50000))
```

```python
plt.figure(figsize=(5,5))
plt.title('Distribution of mileage')
sns.histplot(data, x='Transmission',hue='Engine Type',shrink=0.5,alpha=0.5,multiple='dodge',palette={'Petrol':'grey','Diesel':'black'})
```

```python
trans=data.groupby(['Transmission']).median(numeric_only=False).reset_index()
plt.figure(figsize=(5,3))
plt.title('Distribution of mileage')
sns.barplot(data=trans, x='Transmission',y='Mileage',alpha=0.5,palette={'Automatic':'green','Manual':'black'})
```
