You are given a real estate dataset.

The data is located in the file: 'real_estate_price_size.csv'. 

You are expected to create a simple linear regression (similar to the one in the lecture), using the new data.

In this exercise, the dependent variable is 'price', while the independent variables is 'size'.

Good luck!

## Import the relevant libraries

```python
import pandas as pd #import data that is stored locally
import matplotlib.pyplot as plt #used for ploting
import statsmodels.api as sm #all the models in statistics
from statsmodels.formula.api import ols
```

## Load the data


```python
data= pd.read_csv('real_estate_price_size.csv')
```

```python
data
```

## Create the regression

### Declare the dependent and the independent variables

```python
#dependent_variable 
y = data['price'] 
#independent_variable 
x1 = data['size']
```

### Explore the data

```python
plt.scatter(x1,y,c="blue",label='Data' )
plt.xlabel("Size", fontsize=18)
plt.ylabel("Price", fontsize=18)
plt.title('Regression Analysis', fontsize=20)
plt.legend()
plt.show()
```

### Regression itself

```python
x=sm.add_constant(x1)
model=sm.OLS(y,x).fit()
results=model.summary()
results
```

### Plot the regression line on the initial scatter

```python
plt.scatter(x1,y,c="blue",label='Data' )
yhat=1.019e+05+223.1787*x1
plt.plot(x1,yhat,c="red",label='Regression Line')
plt.xlabel("Size", fontsize=18)
plt.ylabel("Price", fontsize=18)
plt.title('Regression Analysis', fontsize=20)
plt.legend()
plt.show()
```
